# Test Dataset

A generated dataset for testing purposes.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

N = 200000
X, _ = make_blobs(n_samples=N, n_features=2, centers=[[0, 0]], cluster_std=3, random_state=2)
means = X.mean(axis=0)

df = pd.DataFrame(X, columns=["A", "B"])

clustering = DBSCAN(eps=0.32, min_samples=100).fit(df[["A", "B"]])
df["Y"] = clustering.labels_
df.to_csv("./testData.csv", header=False, index=False, sep=";")

df["Y"].hist()

df.plot.hexbin(
  x="A", 
  y="B",
  reduce_C_function=np.sum,
  gridsize=100,
  cmap="viridis",
  figsize=[7, 6],
)
```
